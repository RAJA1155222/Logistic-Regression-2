Q1. What is the purpose of Grid Search CV in machine learning, and how does it work?

Grid Search CV is used to find the best hyperparameter values for a model by systematically trying all combinations from a predefined grid and evaluating each one using cross-validation.

How it works:

You define a grid of hyperparameters.

For each combination:

Perform k-fold cross-validation.

Compute the average validation score.

Select the combination with the best mean CV score.

In practice, this is commonly done using tools provided by
scikit-learn.

Q2. Difference between Grid Search CV and Randomized Search CV, and when to use which?
Aspect	Grid Search CV	Randomized Search CV
Search strategy	Tries all combinations	Tries random combinations
Computation	Expensive for large grids	Much faster
Coverage	Exhaustive	Approximate
Best when	Few hyperparameters and small ranges	Large search space or limited time

When to choose

Use Grid Search CV when the number of combinations is small.

Use Randomized Search CV when:

the parameter space is large,

or training is expensive.

Q3. What is data leakage, and why is it a problem? Give an example.

Data leakage occurs when information from the test set (or future data) is used during training.

Why it is a problem:

It leads to over-optimistic performance estimates.

The model performs poorly in real-world deployment.

Example

If you compute feature scaling (mean and standard deviation) using the entire dataset before train–test split, information from the test data leaks into training.

Q4. How can you prevent data leakage?

Common best practices:

Always perform:

train–test split before any preprocessing.

Fit preprocessing steps only on training data.

Use pipelines so that:

scaling, imputation and model fitting are applied correctly inside cross-validation.

Never use future or target-related information in feature engineering.

Q5. What is a confusion matrix, and what does it tell you?

A confusion matrix is a table that summarizes the predictions of a classification model.

For binary classification:

	Predicted Positive	Predicted Negative
Actual Positive	TP	FN
Actual Negative	FP	TN

It tells you:

how many samples were classified correctly,

and what kinds of errors the model is making.

Q7. How can you interpret a confusion matrix to understand the types of errors?

Look at:

False Positives (FP)
→ model predicts positive when the actual class is negative.

False Negatives (FN)
→ model predicts negative when the actual class is positive.

If:

FP is large → the model is generating many false alarms.

FN is large → the model is missing many real positive cases.

This directly shows the nature of the model’s mistakes.

Q10. How can a confusion matrix help identify biases or limitations in a model?

A confusion matrix can reveal:

very high FN for one class
→ the model is biased toward predicting the other class.

very high FP for a certain class
→ the model over-predicts that class.

For example:

If a medical classifier shows many FN for the positive (disease) class, it indicates a serious limitation: the model is systematically missing real cases.

If one class is almost always predicted correctly and the other is not, the model is likely biased due to:

class imbalance,

poor feature representation,

or an unsuitable decision threshold.

Hence, the confusion matrix helps diagnose class-specific weaknesses, not just overall performance.
